```python
df_economic = pd.read_csv("economic_data.csv")
```

```python
paises_disponiveis = ['australia', 'canada', 'china', 'europe', 'japan', 'new zealand', 'switzerland', 'united kingdom', 'united states']
```

```python
paises_pesquisa = df_economic[df_economic['country'].isin(paises_disponiveis)]
```

```python
df_desemprego_disponiveis = paises_pesquisa[['country', 'unemployment rate']]
```

```python
df_desemprego_disponiveis = df_desemprego_disponiveis.drop_duplicates()
```

```python
df_desemprego_disponiveis.reset_index(drop=True, inplace=True)
print(df_desemprego_disponiveis)
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df_desemprego_disponiveis['country'], df_desemprego_disponiveis['unemployment rate'], color='pink')
plt.xlabel('País')
plt.ylabel('taxa de desemprego (%)')
plt.title('Taxa de desemprego por país')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def calcular_media_pmi(df):
    media_pmi = df['manufacturing pmi'].mean()
    return media_pmi
```

```python
def visualizar_distribuicao_pmi(df):
    plt.figure(figsize=(10, 6))
    plt.plot(df['Date'], df['manufacturing pmi'], marker='o', linestyle='-')
    plt.xlabel('Date')
    plt.ylabel('Manufacturing PMI')
    plt.title('Manufacturing PMI Distribution over Time')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
