```python
import csv
import pandas as pd
```

```python
import sqlite3
```

```python
file = open("economic_data.csv")
```

```python
conteudo = csv.reader(file)
```

```python
connection = sqlite3.connect("banco_de_dados_economic_data.db")
```

```python
cursor = connection.cursor()
```

```python
# CRIAR TABELA 

cursor.execute(''' CREATE TABLE paises (
    country TEXT,
    date TEXT,
    manufacturing_pmi REAL,
    services_pmi REAL,
    consumer_confidence REAL,
    interest_rates REAL,
    cpi_yoy REAL,
    core_cpi REAL,
    unemployment_rate REAL,
    gdp_yoy REAL,
    ticker TEXT,
    open REAL,
    high REAL,
    low REAL,
    close REAL)''') 
```

```python

inserir_conteudo = "INSERT INTO paises(country, date, manufacturing_pmi, services_pmi, consumer_confidence, interest_rates, cpi_yoy, core_cpi, unemployment_rate, gdp_yoy, ticker, open, high, low, close) VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)"
```

```python
cursor.executemany(inserir_conteudo, conteudo) 
```

```python
selecionar_tudo = "SELECT * FROM paises"
```

```python
entradas = cursor.execute(selecionar_tudo).fetchall()
```

```python
for entrada in entradas:
    print(entrada)
```

```python
connection.commit()
```

```python
connection.close
```
